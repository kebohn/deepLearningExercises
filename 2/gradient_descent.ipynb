```python
import numpy
import matplotlib.pyplot as plt
```

```python
def compute_loss(w_1, w_2):
    return w_1 ** 2 + w_2 ** 2 + 30 * numpy.sin(w_1) * numpy.sin(w_2)

def compute_gradient(w):
    return numpy.array([2 * w[0] + 30 * numpy.sin(w[1]) * numpy.cos(w[0]), 2 * w[1] + 30 * numpy.sin(w[0]) * numpy.cos(w[1])])

def gradient_descent(w, theta, iter):
    for _ in range(iter+1):
        gradient = compute_gradient(w)
        w -= theta * gradient
    return w
```

```python
# run gradient descent
theta = 0.01
weights = numpy.random.uniform(-10., 10., 2)
optimized_weights = gradient_descent(numpy.copy(weights), theta, 1000)

# init weights
w_1 = numpy.arange(-10., 10., 0.1)
w_2 = numpy.arange(-10., 10.,0.1)
xx, yy = numpy.meshgrid(w_1,w_2)

# compute loss value
zz = compute_loss(xx,yy)

# surface plot
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d',azim=-60, elev=20)
ax.plot_surface(xx, yy, zz, cmap="jet", alpha=.1)

# plot start and end point
ax.plot(weights[0], weights[1], compute_loss(weights[0], weights[1]), "X", c="red")
ax.plot(optimized_weights[0], optimized_weights[1], compute_loss(optimized_weights[0], optimized_weights[1]), "X", c="blue")
```

```python
print(f"Weights = {weights[0], weights[1]}")
print(f"Loss: {compute_loss(weights[0], weights[1]): .4}") 
```

```python
print(f"Optimized weights = {optimized_weights[0], optimized_weights[1]}")
print(f"Optimized Loss: {compute_loss(optimized_weights[0], optimized_weights[1]): .4}") 
```
