```python
import numpy
import matplotlib.pyplot as plt
```

```python
def compute_loss(w_1, w_2):
    return w_1 ** 2 + w_2 ** 2 + 30 * numpy.sin(w_1) * numpy.sin(w_2)

def compute_gradient(w):
    return numpy.array([2 + 30 * numpy.cos(w[0]), 2 + 30 * numpy.cos(w[1])])

def gradient_descent(w, theta, iter):
    w_copy = numpy.copy(w)
    for _ in range(iter+1):
        gradient = compute_gradient(w_copy)
        w_copy -= theta * gradient
    return w_copy
```

```python
# run gradient descent
theta = 0.2
weights = numpy.random.rand(2)
optimized_weights = gradient_descent(weights, theta, 100)

# init weights
w_1 = numpy.arange(-10., 10., 0.1)
w_2 = numpy.arange(-10., 10.,0.1)
xx, yy = numpy.meshgrid(w_1,w_2)

# compute loss value
compute_loss_vectorized = numpy.vectorize(compute_loss)
zz = compute_loss_vectorized(xx,yy)

# surface plot
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d',azim=-40, elev=50)
ax.plot_surface(xx, yy, zz, cmap="jet", alpha=.8)

# plot start end end point
ax.plot(weights[0], weights[1], compute_loss(weights[0], weights[1]), "kx")
```
